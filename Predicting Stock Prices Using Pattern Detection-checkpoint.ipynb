```python
# Imports
import yfinance as yf
import pandas as pd
from statistics import stdev
import matplotlib.pyplot as plt
import datetime
%matplotlib inline
```

## Retrieve and Format Data

* prediction_time_to_interval: used to format times in a way that yfinance will accept.

* pull_data: pulls stock information from the yahoo finance servers.

* format_data: Takes raw data from yahoo and removes unecessary data.  Also determines percent change over the time periods.

* determine_past_values: determines similarity of past instances by using r^2 

* closest_past_values: pulls instances with lowest r^2 values

```python
def prediction_time_to_interval(pt):
    if   pt == '1d': return '1d'
    elif pt == '1m': return '1mo'
    elif pt == '1yr': return '1mo'
    else: return -1

def pull_data(stockTicker, predictionTime):
    stockData = yf.Ticker(stockTicker)
    return stockData.history(period='max', 
                             interval=prediction_time_to_interval(predictionTime))

def format_data(rawData):
    # Add column for simple percentage change over a time period
    rawData['percent_change'] = 2 * (rawData['Close'] - rawData['Open']) / (rawData['Close'] + rawData['Open'])
    # only return useful columns
    return rawData[['Open', 'Close', 'percent_change']].copy()

def determine_past_values(stockData, precision):
    goalValues = list(stockData['percent_change'])[-precision:]
    sumValues = [0]*len(stockData)
    nextChange = []
    for i in range(len(stockData) - precision - 1):
        sectionValues = list(stockData.iloc[i:i+precision]['percent_change'])
        for j in range(precision):
            sumValues[i] += (sectionValues[j] - goalValues[j])**2
            
        nextChange.append(stockData.iloc[i+precision+1]['percent_change'])
    
    resultsDB = pd.DataFrame({'difference':sumValues[:len(nextChange)], 'next_change':nextChange})
    return resultsDB

def closest_past_values(stockData, precision, options):
    prevVals = determine_past_values(stockData, precision)
    return prevVals.sort_values(by=['difference'])[:options]
    
```

## Prediction and Graphing

* predict_value: used to tie tools together, gives raw data for use later
* graph: makes a graph from given data
* pretty_predict: gives some information and calls the graph

```python
def predict_value(stockTicker, predictionTime, precision, options):
    # pull info for stockTicker
    stockData = pull_data(stockTicker, predictionTime)
    stockData = stockData.dropna()
    stockData = format_data(stockData)

    # find patterns
    previous = closest_past_values(stockData, precision, options)
    
    # make prediction
    return stockData, previous

def graph(ticker, history, predictionTime, guesses):
    labels = ['worst case', 'best case', 'average case', 'weighed case']
    dd = 1 if predictionTime == '1d' else 30
    titleText = "Predicted Stock Value for " + ticker.upper() + " for next "
    titleText += "business day." if dd==1 else "month."
    
    for i in range(len(labels)):
        guessSeries = history.append(pd.Series({datetime.datetime.today() + datetime.timedelta(days=dd): guesses[i]}))
        guessSeries[-7:].plot(label=labels[i])
    plt.legend()
    plt.title(titleText)
    plt.grid(True)

def pretty_predict(stockTicker, predictionTime, precision, options):
    dataset, predictions = predict_value(stockTicker, predictionTime, precision, options)
    lastVal = dataset['Close'][-1]
    averageGuess = lastVal * (1 + predictions['next_change'].mean())
    worstGuess = lastVal * (1 + min(predictions['next_change']))
    bestGuess = lastVal * (1 + max(predictions['next_change']))
    
    print('last:', lastVal)
    print('av:', averageGuess)
    print('worst:', worstGuess)
    print('bestGuess:', bestGuess)
    
    inverted = predictions['difference']**-1
    predictions['weighted'] = inverted/sum(inverted)*predictions['next_change']
    weightedGuess = lastVal * (1 + sum(predictions['weighted']))
    print('weighted:', weightedGuess)
    
    history = dataset['Close']
    guesses = [worstGuess, bestGuess, averageGuess, weightedGuess]
    
    graph(stockTicker, history, predictionTime, guesses)
    
    return stockTicker, history, guesses
    
```

## pretty_predict

* ticker - The name of the stock (coins use a different system)

* predictionTime - 1d or 1m in the future

* precision - how many days are used to build a pattern O(n^2)

* options - amount of matched patterns used to make predictions
